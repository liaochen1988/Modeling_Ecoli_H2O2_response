# Hclustered correlation matrix

```python
def plot_corr(df, size=10, is_plot=0, filename=''):
    '''Plot a graphical correlation matrix for a dataframe.

    Input:
        df: pandas DataFrame
        size: vertical and horizontal size of the plot'''

    # Compute the correlation matrix for the received dataframe
    corr = df.corr() # pearson
    
    # Plot the correlation matrix
    fig, ax = plt.subplots(figsize=(size, size))
    plt.imshow(corr, cmap='RdYlGn_r')
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90);
    plt.yticks(range(len(corr.columns)), corr.columns);
    ax.xaxis.tick_top()
    
    # Add the colorbar legend
    cbar = plt.colorbar(ticks=[-1, 0, 1], aspect=40, shrink=.8)
    plt.clim(-1,1)
    
    if is_plot:
        fig.savefig(filename, dpi=600)
```

```python
metPA=pd.read_excel('../data/Normalized_annotated_metabolomics_ratio2PA14_tidy.xlsx')

# select metabolaites with confidence level >0
metInfo=pd.read_excel('../data/identified_compounds_id_name_v1.xlsx')
conf_index = metInfo['Confidence_level'] > 0.5
metPA_conf = pd.DataFrame(metPA.values[:,conf_index])
metPA_conf.index = metPA.index
#metPA_conf.columns = [metInfo['Confirmed_print_name'][i] for i in range(len(metPA.columns)) if conf_index[i]]
metPA_conf.columns = [metInfo['Metabolic_identifier'][i] for i in range(len(metPA.columns)) if conf_index[i]]

X = metPA_conf.corr().values # pearson
d = sch.distance.pdist(X,'cityblock')
L = sch.linkage(d, method='average')
ind = sch.fcluster(L, 0*d.max(), 'distance')
columns = [metPA_conf.columns.tolist()[i] for i in list((np.argsort(ind)))]
metPA_conf = metPA_conf.reindex(columns, axis=1)

plot_corr(metPA_conf, size=18, is_plot=True, filename = '../data/normalized_metabolomics_hcluster_corr_mat_curated.pdf')
```

```python
def plot_heatmap_metabolic(met_id):
    
    metPA=pd.read_excel('../data/Normalized_annotated_metabolomics_ratio2PA14_tidy.xlsx')
    peakArea = metPA[met_id]

    nmut = int(len(peakArea)/3)
    peakArea_array = np.zeros(shape=(3,nmut))
    mutant_id = []

    count = 0
    for col in range(nmut):
        mutant_id.append(peakArea.index[col].split('_')[0])
        peakArea_array[0,col] = peakArea[col+count]
    
    count += nmut
    for col in range(nmut):
        peakArea_array[1,col] = peakArea[col+count]
    
    count += nmut
    for col in range(nmut):
        peakArea_array[2,col] = peakArea[col+count]
    
    df = pd.DataFrame(np.log10(peakArea_array))
    fig = plt.gcf()
    fig.set_size_inches(nmut,3)
    sns.heatmap(df, cmap = 'RdYlBu_r', yticklabels=['R1','R2','R3'], xticklabels=mutant_id)
    #plt.show()
    
    return peakArea_array
```

```python
def plot_subgroup_correlation(group):
    metPA=pd.read_excel('../data/Normalized_annotated_metabolomics_ratio2PA14_tidy.xlsx')
    metPA=metPA[group]

    # select metabolaites with confidence level >0
    metInfo=pd.read_excel('../data/identified_compounds_id_name_v1.xlsx')
    index = [metInfo[metInfo['Metabolic_identifier']==i].index for i in group]
    new_column_names=[]
    for i in range(len(group)):
        new_column_names.append(metInfo['Confirmed_print_name'][index[i]].tolist()[0])
    metPA.columns = new_column_names
    
    corr = metPA.corr()

    # Plot the correlation matrix
    fig, ax = plt.subplots(figsize=(len(group), len(group)))
    plt.imshow(corr, cmap='RdYlGn_r')
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90, fontsize=15);
    plt.yticks(range(len(corr.columns)), corr.columns, fontsize=15);
    ax.xaxis.tick_top()
    
    # Add the colorbar legend
    cbar = plt.colorbar(ticks=[-1, 0, 1], aspect=40, shrink=.8)
    plt.clim(-1,1)
```
